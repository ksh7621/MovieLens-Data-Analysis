```python
import time
import operator

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
#영화 이름 - u.item
#성별 직업 = u.user
#장르-u.item

data_path = '/Users/ksh76/ml-100k/u.data'
item_path = '/Users/ksh76/ml-100k/u.item'
user_path = '/Users/ksh76/ml-100k/u.user'
occupation_path = '/Users/ksh76/ml-100k/u.occupation'
genre_path = '/Users/ksh76/ml-100k/u.genre'

rating_data = pd.io.parsers.read_csv(data_path,names=['user_id', 'movie_id', 'rating', 'time'], delimiter='\t')

movie_data = pd.io.parsers.read_csv(item_path, 
                                    names=['movie_id', 'title', 'release_date','video_realease_date', 'url', 'unknown', 'Action'
                                          ,'Adventure','Animation',"Children's",'Comedy','Crime','Documentary','Drama'
                                           ,'Fantasy','Film-Noir','Horror','Musical','Mystery','Romance','Sci-Fi','Thriller'
                                          ,'War','Western'], delimiter='|')

user_data = pd.io.parsers.read_csv(user_path, names=['user_id', 'age', 'gender', 'occupation', 'zipcode'], delimiter='|')

occupation_data = pd.io.parsers.read_csv(occupation_path, names=['occupation'], delimiter='\n')

genre_data = pd.io.parsers.read_csv(genre_path, names=['genre', 'num'], delimiter='|')

pd.set_option('display.max_row',100)
pd.set_option('display.max_columns',100)
```

```python
data = pd.merge(pd.merge(rating_data, user_data), movie_data)
```

# 2_A

#### Make a list of the top 100 most popular movies_Extract a list of movies with the highest number of ratings


```python
#for all users 
rating_all = pd.DataFrame(data.groupby('title').size().to_frame('size'))
rating_all.sort_values(by='size', ascending=False).head(100)
```

```python
#by female
rating_gender_female = data.groupby('gender').get_group('F')
rating_female = pd.DataFrame(rating_gender_female.groupby('title').size().to_frame('size'))
rating_female.sort_values(by='size', ascending=False).head(100)
```

```python
# by male 
rating_gender_male = data.groupby('gender').get_group('M')
rating_male = pd.DataFrame(rating_gender_male.groupby('title').size().to_frame('size'))
rating_male.sort_values(by='size', ascending=False).head(100)
```

```python
# by occupation
occ = occupation_data['occupation']

occ_list = []
for i in range(occ.shape[0]):
    occ_list.append(occ[i])
for i in range(occ.shape[0]):
    rating_occupation = data.groupby('occupation').get_group(occ_list[i])
    print("\noccupation:", occ_list[i])    
    rating_occ = pd.DataFrame(rating_occupation.groupby('title').size().to_frame('size'))    
    print(rating_occ.sort_values(by='size', ascending=False).head(100))  
```

```python
#for all users
gen = genre_data['genre']

gen_list = []
for i in range(gen.shape[0]):
    gen_list.append(gen[i])

genre_pd = pd.DataFrame({'genre': list([gen_list[i] for i in range(gen.shape[0])]), 
              "size":list([data.groupby(gen_list[i]).size()[1] for i in range(gen.shape[0])])
             })
genre_pd.sort_values(by='size', ascending=False)
```

```python
#occupation

for con in range(occ.shape[0]):
    if (occ_list[con] == 'administrator' or occ_list[con] == 'artist' or occ_list[con] == 'doctor' or occ_list[con] == 'healthcare'
    or occ_list[con] == 'lawyer' or occ_list[con] == 'librarian' or occ_list[con] == 'marketing' 
    or occ_list[con] == 'retired' or occ_list[con] == 'salesman' or occ_list[con] == 'scientist'):
        genre_rating_occupation = data.groupby('occupation').get_group(occ_list[con])
        print("\n occupation: ",occ_list[con])        
        genre_rating_occupation_pd = pd.DataFrame({'genre': list([gen_list[i] for i in range(gen.shape[0])]), 
              "size":list([genre_rating_occupation.groupby(gen_list[j]).size()[1] if gen_list[j] != 'unknown' 
                           else genre_rating_occupation.groupby(gen_list[j]).size()[0] - genre_rating_occupation.groupby(gen_list[j]).size()[0]
                           for j in range(gen.shape[0])])})        
        print(genre_rating_occupation_pd.sort_values(by='size', ascending=False)) 
    elif(occ_list[con] == 'homemaker'):        
        genre_rating_occupation = data.groupby('occupation').get_group(occ_list[con])
        print("\n occupation: ",occ_list[con])        

        genre_rating_occupation_pd = pd.DataFrame({'genre': list([gen_list[i] for i in range(gen.shape[0])]), 
              "size":list([genre_rating_occupation.groupby(gen_list[j]).size()[1] 
                           if gen_list[j] != 'unknown' and gen_list[j] != 'Western' else genre_rating_occupation.groupby(gen_list[j]).size()[0] - genre_rating_occupation.groupby(gen_list[j]).size()[0] 
                           for j in range(gen.shape[0])])}) 
        print(genre_rating_occupation_pd.sort_values(by='size', ascending=False))
    else:
        genre_rating_occupation = data.groupby('occupation').get_group(occ_list[con])
        print("\n occupation: ",occ_list[con])        
        genre_rating_occupation_pd = pd.DataFrame({'genre': list([gen_list[i] for i in range(gen.shape[0])]), 
              "size":list([genre_rating_occupation.groupby(gen_list[j]).size()[1] for j in range(gen.shape[0])])})        
        print(genre_rating_occupation_pd.sort_values(by='size', ascending=False))
    
        
```
